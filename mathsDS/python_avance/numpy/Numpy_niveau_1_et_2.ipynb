```python
import pandas as pd
import numpy as np

df = pd.read_csv('./distance_villes.csv', sep=',', index_col=0, skipfooter=3)
mat = df.to_numpy()
display(df)
```

```python
columns_to_skip = 'Distance (kilomètres)'
df_ct = list(pd.read_csv('./distance_villes.csv', sep=',', index_col=False, skipfooter=3, usecols=lambda x: x not in columns_to_skip))
display(df_ct)
```

2) Remplacez les valeurs manquantes par une valeur cohérente

```python
np.isnan(mat)
df.fillna(0)
```

```python
mat = np.where(np.isnan(mat), 0, mat)
mat = np.nan_to_num(mat)
mat
```

```python
mxm = np.max(mat[np.isfinite(mat)])
mat[np.isinf(mat)] = mxm
mat
```

3) Quelle est la meilleure ville où habiter si on doit voyager régulièrement sur toutes les villes du jeu de donnée ? 
- utiliser numpy (et pas des boucles)
- on suppose que la meilleure est celle qui permet un temps de trajet moyen le plus court

```python
ct = np.where(np.mean(mat, axis=0) == np.amin(np.mean(mat, axis=0)))
df_ct[ct[0][0]]
```

4) Quelle est la meilleure ville avoir des trajets le plus court possible ? quel est alors le pire temps de trajet ?

```python
dist_min = np.where(np.amax(mat, axis=0) == np.amin(np.amax(mat, axis=0)))
df_ct[dist_min[0][0]]
```

```python
dist_min = np.where(np.average(mat, axis=0) == np.amax(np.average(mat, axis=0)))
df_ct[dist_min[0][0]]
```
